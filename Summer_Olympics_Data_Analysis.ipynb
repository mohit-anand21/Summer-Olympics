# Summer Olympics Data Analysis Assignment

### Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Reading Summer Olympics Dataset

```python
df = pd.read_csv("summer.csv", index_col = 0)
df.head()
```

## 1. In how many cities Summer Olympics is held so far?

```python
print("Summer Olympics has been held at \"", df.City.nunique(), "\" different countries so far!!", sep='')
```

## 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
# Creating a dataframe based on how many Gold Medals are won w.r.t Sports
df1 = pd.DataFrame(df[df.Medal  == 'Gold'].groupby('Sport')['Medal'].agg('count').sort_values(ascending=False).reset_index())

# Printing the sport having most number of Gold Medals
print("\nThe sport having the highest number of Gold Medals is \"", df1.Sport[0], "\" with a total of ", 
      df1.Medal[0], " medals so far!!", sep='')

# Plotting Top 5 Sports which have most number of Gold Medals
plt.figure(figsize=(10,7))
plots = sns.barplot(x='Sport',y='Medal',data=df1.head(), palette='Spectral_r', edgecolor='black', lw=2)
plt.xlabel('Top 5 Sports',size=15)
plt.ylabel('Total Number of Gold Medals',size=15)
plt.title('Top 5 Sports with highest number of Gold Medals',size=15)
for bar in plots.patches:
    plt.annotate(format(bar.get_height(), '.0f'), (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                 ha='center', va='center', size=12, xytext=(0, 8), textcoords='offset points')
```

## 3. Which sport is having most number of medals so far? (Top 5)

```python
# Creating a dataframe based on how many Medals are won w.r.t Sports
df2 = pd.DataFrame(df.groupby('Sport')['Medal'].agg('count').sort_values(ascending=False)).reset_index()

# Printing the Sport having the most number of medals
print("\nThe sport having the highest number of medals so far is \"", df2.Sport[0], "\"!", sep='')

# Plotting the Top 5 Sports having the most number of medals
plt.figure(figsize=(10,7))
plots = sns.barplot(x='Sport',y='Medal',data=df2.head(), palette='hot', edgecolor='black', lw=2)
plt.xlabel('Top 5 Sports',size=15)
plt.ylabel('Total Number of Medals',size=15)
plt.title('Top 5 Sports with highest number of Medals',size=15)
for bar in plots.patches:
    plt.annotate(format(bar.get_height(), '.0f'), (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                 ha='center', va='center', size=12, xytext=(0, 8), textcoords='offset points')
```

## 4. Which player has won most number of medals? (Top 5)

```python
# Creating a dataframe based on how many Medals are won by an individual Player
df3 = pd.DataFrame(df.groupby('Athlete')['Medal'].agg('count').sort_values(ascending=False)).reset_index()

# Printing the player name who won the most number of medals
name = df3.Athlete[0]
name = name.split(', ')
print("\n\"", name[1], " ", name[0].capitalize() , "\" has won the most number of medals so far in Summer Olympics!", sep='')

# Plotting the Top 5 players who won the highest number of medals
plt.figure(figsize=(11,7))
plots = sns.barplot(x='Athlete',y='Medal',data=df3.head(), palette='coolwarm_r', edgecolor='black', lw=2)
plt.xlabel('Top 5 Sportsperson',size=15)
plt.ylabel('Total Number of Medals',size=15)
plt.title('Top 5 Sportsperson with highest number of Medals',size=15)
for bar in plots.patches:
    plt.annotate(format(bar.get_height(), '.0f'), (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                 ha='center', va='center', size=12, xytext=(0, 8), textcoords='offset points')
```

## 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
# Creating a dataframe based on how many Gold Medals are won by an individual player
df4 = pd.DataFrame(df[df.Medal == 'Gold'].groupby('Athlete')['Medal'].agg('count').sort_values(ascending=False)).reset_index()

# Printing the player name who has won the most number of Gold Medals
name1 = df4.Athlete[0]
name1 = name1.split(', ')
print("\n\"", name1[1], " ", name1[0].capitalize() , "\" has won the most number of medals so far in Summer Olympics!", sep='')

# Plotting the Top 5 players who has won the highest number of Gold Medals
plt.figure(figsize=(10,7))
plots = sns.barplot(x='Athlete',y='Medal',data=df4.head(), palette='autumn', edgecolor='black', lw=2)
plt.xlabel('Top 5 Sportsperson',size=15)
plt.ylabel('Total Number of Gold Medals',size=15)
plt.title('Top 5 Sportsperson with highest number of Gold Medals',size=15)
for bar in plots.patches:
    plt.annotate(format(bar.get_height(), '.0f'), (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                 ha='center', va='center', size=12, xytext=(0, 8), textcoords='offset points')
```

## 7. Which event is most popular in terms on number of players? (Top 5)

```python
# Creating a dataframe based on how many Events are most popular w.r.t number of players
df5 = pd.DataFrame(df.groupby('Event')['Athlete'].agg('count').sort_values(ascending=False)).reset_index()

# Printing the Event name which is most popular in terms of numer of players
print("\nThe event that is most popular in terms of number of players is \"", df5.Event[0], "\".", sep='')

# Plotting the Top 5 most popular Events in terms of number of players
plt.figure(figsize=(10,7))
plots = sns.barplot(x='Event',y='Athlete',data=df5.head(), palette='vlag_r', edgecolor='black', lw=2)
plt.xlabel('Top 5 Event',size=15)
plt.ylabel('Total Number of Players',size=15)
plt.title('Top 5 most popular Events in terms of number of players',size=15)
for bar in plots.patches:
    plt.annotate(format(bar.get_height(), '.0f'), (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                 ha='center', va='center', size=12, xytext=(0, 8), textcoords='offset points')
```

## 8. Which sport is having most female Gold Medalists? (Top 5)
